```python
import numpy as np
from astropy.modeling.blackbody import blackbody_nu, blackbody_lambda
import astropy.units as u
import matplotlib.pyplot as plt
```

```python
bb_star = blackbody_lambda(np.logspace(3, 7), 10000).to(u.W / (u.m* u.m *u.steradian* u.m))
bb_dust = blackbody_lambda(np.logspace(3, 7), 100).to(u.W / (u.m* u.m *u.steradian* u.m))
```

```python
plt.figure(figsize=(10, 8))
plt.loglog(np.logspace(3, 7), bb_star, label='Star at 1e4 K')
plt.loglog(np.logspace(3, 7), bb_dust, label='Dust at 100 K')

plt.ylim([1e-3, 1e15])
plt.legend(loc='best', fontsize=15)

plt.ylabel('B$_{\lambda}$ [W str$^{-1}$ m$^{-3}$]', fontsize=25)
plt.xlabel('Wavelength [$\AA$]', fontsize=25)
```

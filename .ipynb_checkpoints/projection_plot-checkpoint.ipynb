```python
ds.index
ad = ds.all_data()
```

```python
obj = caesar.load('/Volumes/orange/narayanan/desika.narayanan/gizmo_runs/simba/m25n512/output/Groups/caesar_0305_z0.000.hdf5')
```

```python
center = obj.galaxies[93].pos.in_units("kpc")
width = ds.quan(200, 'kpccm')
size = [width/2, width/2, width/2]
```

```python
image_array = yt.off_axis_projection(
    ds.box(center - size, center + size),
    center,
    normal_vector=[np.cos(0), np.sin(0), 0.0],
    width = ds.arr([500.0, 500.0, 500.0], 'kpccm'),
    resolution=[720, 720],
    item=("gas", "density"),
    north_vector=[0.0, 0.0, 1.0],
)
```

```python
plt.imshow(
    image_array.value,
    cmap='magma',
    norm=matplotlib.colors.LogNorm(),                                  
)
```

```python
density = ad[("PartType4","Masses")]
wdens = np.where(density == np.max(density))
coordinates = ad[("PartType4","Coordinates")]
center = coordinates[wdens][0]
print ('center = ',center)
```

```python
new_box_size = ds.quan(70,'code_length')
```

```python
px = yt.ProjectionPlot(ds, 'x', ("gas","density"), center=center, width=new_box_size, resolution=[400, 400])
px.set_background_color(("gas","density"))
px.show()
```

```python
c = [0.5, 0.5, 0.5]

# Our image plane will be normal to some vector.  For things like collapsing
# objects, you could set it the way you would a cutting plane -- but for this
# dataset, we'll just choose an off-axis value at random.  This gets normalized
# automatically.
L = [0.0, 1.0, 0.0]

# Our "width" is the width of the image plane as well as the depth.
# The first element is the left to right width, the second is the
# top-bottom width, and the last element is the back-to-front width
# (all in code units)
normal_vector=[np.cos(0), np.sin(0), 0.0]
width = ds.arr([100.0, 100.0, 100.0], 'kpccm')

# The number of pixels along one side of the image.
# The final image will have Npixel^2 pixels.
Npixels = 512

# Create the off axis projection.
# Setting no_ghost to False speeds up the process, but makes a
# slightly lower quality image.
image = yt.off_axis_projection(ds, center, L, width, resolution=[512, 512], item=("gas", "density"), no_ghost=False)
```

```python
import fsps
import numpy as np
import matplotlib.pyplot as plt
```

```python
sp = fsps.StellarPopulation(compute_vega_mags=False, zcontinuous=1, 
                            sfh=3, dust_type=2, dust2=1.0, tage=13.7)
sp.set_tabular_sfh(age=np.array([0.0, 13.7]), sfr=np.array([10., 10.]))
```

```python
#At 10 Myr
wav_10, spec_10 = sp.get_spectrum(tage=0.01, peraa=True)
plt.figure(figsize=(10, 8))
plt.plot(wav_10,spec_10) #spec is L_sun / Angstrom
plt.loglog()
plt.ylim([1e2, 1e10])
plt.ylabel(r'$\frac{L_{\odot}}{\AA}$', fontsize=20)
plt.xlabel('$\lambda$ [$\AA$]', fontsize=15)


ltot_10 = np.trapz(spec_10, wav_10)

ltot_ergs_10 = ltot_10 * 3.839e33 #to ergs/s

print('Bolometric Luminosity =',ltot_10,'L_sun.')
print('Log(L_tot / SFR) = ', np.log10(ltot_ergs_10 / sp.sfr))

```

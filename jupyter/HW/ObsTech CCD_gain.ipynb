```python
exp01 = np.empty([exp01_selection1.shape[0], exp01_selection1.shape[1], 3])
exp5 = np.empty([exp5_selection1.shape[0], exp5_selection1.shape[1], 3])
exp10 = np.empty([exp10_selection1.shape[0], exp10_selection1.shape[1], 3])
exp30 = np.empty([exp30_selection1.shape[0], exp30_selection1.shape[1], 3])
exp60 = np.empty([exp60_selection1.shape[0], exp60_selection1.shape[1], 3])
exp100 = np.empty([exp100_selection1.shape[0], exp100_selection1.shape[1], 3])
```

```python
exp01[:,:,0] = exp01_selection1[:,:]
exp01[:,:,1] = exp01_selection2[:,:]
exp01[:,:,2] = exp01_selection3[:,:]

exp5[:,:,0] = exp5_selection1[:,:]
exp5[:,:,1] = exp5_selection2[:,:]
exp5[:,:,2] = exp5_selection3[:,:]


exp10[:,:,0] = exp10_selection1[:,:]
exp10[:,:,1] = exp10_selection2[:,:]
exp10[:,:,2] = exp10_selection3[:,:]

exp30[:,:,0] = exp30_selection1[:,:]
exp30[:,:,1] = exp30_selection2[:,:]
exp30[:,:,2] = exp30_selection3[:,:]

exp60[:,:,0] = exp60_selection1[:,:]
exp60[:,:,1] = exp60_selection2[:,:]
exp60[:,:,2] = exp60_selection3[:,:]

exp100[:,:,0] = exp100_selection1[:,:]
exp100[:,:,1] = exp100_selection2[:,:]
exp100[:,:,2] = exp100_selection3[:,:]
```

```python
#pixel wise mean for each exposure time
exp01_mean = np.ravel(np.mean(exp01, axis=2))
exp5_mean = np.ravel(np.mean(exp5, axis=2))
exp10_mean = np.ravel(np.mean(exp10, axis=2))
exp30_mean = np.ravel(np.mean(exp30, axis=2))
exp60_mean = np.ravel(np.mean(exp60, axis=2))
exp100_mean = np.ravel(np.mean(exp100, axis=2))
```

```python
plt.plot(exp01_mean)
```

```python
#pixel wise rms for each exposure time
exp01_srms = np.ravel(np.std(exp01, axis=2)**2)
exp5_srms = np.ravel(np.std(exp5, axis=2)**2)
exp10_srms = np.ravel(np.std(exp10, axis=2)**2)
exp30_srms = np.ravel(np.std(exp30, axis=2)**2)
exp60_srms = np.ravel(np.std(exp60, axis=2)**2)
exp100_srms = np.ravel(np.std(exp100, axis=2)**2)
```

```python
np.ravel(exp30_srms)
```

```python
var = np.array([exp01_srms, exp5_srms, exp10_srms, exp30_srms, exp60_srms, exp100_srms])
signal = np.array([exp01_mean, exp5_mean, exp10_mean, exp30_mean, exp60_mean, exp100_mean])

var_med = []
sig_med = []
for i in range(0, len(var)):
    var_med.append(np.median(var[i]))
    sig_med.append(np.median(signal[i]))

    
    
var_wo60 = np.array([exp01_srms, exp5_srms, exp10_srms, exp30_srms,  exp100_srms])
signal_wo60 = np.array([exp01_mean, exp5_mean, exp10_mean, exp30_mean, exp100_mean])

   

    
var_med_wo60 = []
sig_med_wo60 = []
for i in range(0, len(var_wo60)):
    var_med_wo60.append(np.median(var_wo60[i]))
    sig_med_wo60.append(np.median(signal_wo60[i]))
```

```python
#for 5s
from scipy.optimize import curve_fit
plt.figure(figsize=(10,8))
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['axes.linewidth'] = 2.5
plt.tick_params('both', length=10, width=2, which='major')
plt.tick_params('both', length=4.5, width=2, which='minor')

labs = ['0.1s', '5s', '10s', '30s', '60s', '100s']
cols = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']


def lin(x, m, b):
    return x*m + b

x_array = np.linspace(0, 60000, 10000)
popt1, pcov1 = curve_fit(lin, sig_med, var_med, p0=[0.5,  2000.])
popt2, pcov2 = curve_fit(lin, sig_med_wo60, var_med_wo60, p0=[0.5,  2000.])

for i in range(0, len(signal)):
    plt.scatter(signal[i], var[i], c=cols[i], label=labs[i])
    plt.scatter(sig_med[i], var_med[i], c='black', marker='+', s=100)
    plt.plot(x_array, lin(x_array, *popt1), color='black', lw=2, ls=':')
    plt.plot(x_array, lin(x_array, *popt2), color='black', lw=2, ls='--')


plt.annotate('Slope w/60s = 8.67 $\pm$ 9.34', xy=(9000, 2700000), fontsize=20)
plt.annotate('Slope w/o 60s = 1.95 $\pm$ 0.27', xy=(9000, 2500000), fontsize=20)


plt.xlabel('Signal (ADU)', fontsize=25)
plt.ylabel('Variance ($RMS^2$)', fontsize=25)
#plt.ylim([0, 810000])
#plt.yscale('log')
#plt.xscale('log')
#plt.xlim([1.5e4, 2e4])

plt.legend(loc='upper right', fontsize=20, frameon=False)
```

```python
slope1 = popt1[0]
perr1 = np.sqrt(np.diag(pcov1))[0]
slope2 = popt2[0]
perr2 = np.sqrt(np.diag(pcov2))[0]

print(slope1, perr1, slope2, perr2)
```

Plots comparing intrinsic and SED derived physical properties

LIR, LFUV, Mdust, Mstar

```python
import matplotlib.pyplot as plt

import pandas as pd
import astropy.constants as const
```

```python
l_sun = const.L_sun
m_sun = const.M_sun
```

```python
#intrinsic

halo401_i = pd.read_csv('/Volumes/SML/UF/cosmo_sim_halos/pd_zooms/401_intrinsic.csv')
halo374_i = pd.read_csv('/Volumes/SML/UF/cosmo_sim_halos/pd_zooms/374_intrinsic.csv')
halo352_i = pd.read_csv('/Volumes/SML/UF/cosmo_sim_halos/pd_zooms/352_intrinsic.csv')
halo287_i = pd.read_csv('/Volumes/SML/UF/cosmo_sim_halos/pd_zooms/287_intrinsic.csv')
```

```python
#derived

halo401_d = pd.read_csv('/Volumes/SML/UF/cosmo_sim_halos/pd_zooms/401_derived.csv')
halo374_d = pd.read_csv('/Volumes/SML/UF/cosmo_sim_halos/pd_zooms/374_derived.csv')
halo352_d = pd.read_csv('/Volumes/SML/UF/cosmo_sim_halos/pd_zooms/352_derived.csv')
halo287_d = pd.read_csv('/Volumes/SML/UF/cosmo_sim_halos/pd_zooms/287_derived.csv')
```

```python
derived_mdust = [halo287_d['Mdust [M_sun]'], halo352_d['Mdust [M_sun]'], halo374_d['Mdust [M_sun]'], halo401_d['Mdust [M_sun]']]
derived_mstar = [halo287_d['Mstar [M_sun]'], halo352_d['Mstar [M_sun]'], halo374_d['Mstar [M_sun]'], halo401_d['Mstar [M_sun]']]
derived_lir = [halo287_d['LIR [L_sun]'], halo352_d['LIR [L_sun]'], halo374_d['LIR [L_sun]'], halo401_d['LIR [L_sun]']]
derived_lfuv = [halo287_d['LFUV [L_sun]'], halo352_d['LFUV [L_sun]'], halo374_d['LFUV [L_sun]'], halo401_d['LFUV [L_sun]']]


intrinsic_mdust = [halo287_i['Mdust'], halo352_i['Mdust'], halo374_i['Mdust'], halo401_i['Mdust']]
intrinsic_mstar = [halo287_i['M_star'], halo352_i['M_star'], halo374_i['M_star'], halo401_i['M_star']] 
intrinsic_lir = [halo287_i['LIR'], halo352_i['LIR'], halo374_i['LIR'], halo401_i['LIR']] / l_sun
intrinsic_lfuv = [halo287_i['LFUV'], halo352_i['LFUV'], halo374_i['LFUV'], halo401_i['LFUV']] / l_sun
```

```python
ids = ['halo287.snapshot110', 'halo352.snapshot130', 'halo374.snapshot115', 'halo401.snapshot115']
colors = ['blue', 'green', 'orange', 'red']
```

```python
fig, ax = plt.subplots(figsize=(10,10),nrows=2, ncols=2)


for i in range(0, len(intrinsic_mdust)):
    ax[0,0].scatter(intrinsic_mdust[i], derived_mdust[i], c=colors[i], label=ids[i])
    ax[0,0].plot([1e5, 1e8],[1e5, 1e8], color='black')
    ax[0, 1].scatter(intrinsic_mstar[i], derived_mstar[i], c=colors[i])
    ax[0,1].plot([1e9, 1e11],[1e9, 1e11], color='black')
    ax[1, 0].scatter(intrinsic_lir[i], derived_lir[i], c=colors[i])
    ax[1,0].plot([1e8, 1e11],[1e8, 1e11], color='black')
    ax[1, 1].scatter(intrinsic_lfuv[i], derived_lfuv[i], c=colors[i])
    ax[1,1].plot([1e7, 1e10],[1e7, 1e10], color='black')
    
ax[0, 0].set_xscale('log')
ax[0, 0].set_yscale('log')
#ax[0, 0].set_ylim([1e5, 1e8])
#ax[0, 0].set_xlim([1e8, 1e8])

ax[0, 1].set_xscale('log')
ax[0, 1].set_yscale('log')
#ax[0, 1].set_ylim([1e9, 1e11])
#ax[0, 1].set_xlim([1e9, 1e11])

ax[1, 0].set_xscale('log')
ax[1, 0].set_yscale('log')
#ax[1, 0].set_ylim([1e8, 1e11])
#ax[1, 0].set_xlim([1e8, 1e11])

ax[1, 1].set_xscale('log')
ax[1, 1].set_yscale('log')
#ax[1, 1].set_ylim([1e14, 1e18])
#ax[1, 1].set_xlim([1e14, 1e18])

ax[0,0].set_ylabel('Derived Mdust')
ax[0,1].set_ylabel('Derived Mstar')
ax[1,0].set_ylabel('Derived LIR')
ax[1, 1].set_ylabel('Derived LFUV')

ax[0,0].set_xlabel('Intrinsic Mdust')
ax[0,1].set_xlabel('Intrinsic Mstar')
ax[1,0].set_xlabel('Intrinsic LIR')
ax[1, 1].set_xlabel('Intrinsic LFUV')

ax[0,0].legend(loc='upper left', fontsize=10)
```

```python
#from coSLED.py example


# Import the despotic library
from despotic import cloud

# Import standard python libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the Milky Way GMC cloud file
gmc = cloud(fileName='/Users/sidneylower/despotic/cloudfiles/MilkyWayGMC.desp')
```

```python
temp_range = np.arange(1.0, 100., 1.) #K
surface_density_range = np.arange(1e20, 1e24, 1e22) #cm^-2
velocity_dispersion_range = np.arange(0., 2e6, 1e5) #cm / s
```

```python
X_varyT = []
for i in temp_range:
    gmc.Tg = i
    #print(gmc.Tg)
    X_varyT.append(gmc.colDen/gmc.lineLum('co')[0]['intTB'])
```

```python
#X_varyT
```

```python
plt.figure(figsize=(10, 8))
plt.plot(temp_range, X_varyT)
plt.ylabel(r'X$_{CO}$ $[\frac{cm^{-2}}{K * km/s}]$', fontsize=25)
plt.xlabel('Gas Temperature [K]', fontsize=23)
plt.yscale('log')
```

```python
X_varyV = []
for i in velocity_dispersion_range:
    gmc.sigmaNT = i
    #print(gmc.Tg)
    X_varyV.append(gmc.colDen/gmc.lineLum('co')[0]['intTB'])
 
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(velocity_dispersion_range / 1e5, X_varyV, s=200)
plt.ylabel(r'X$_{CO}$ $[\frac{cm^{-2}}{K * km/s}]$', fontsize=25)
plt.xlabel('Velocity Dispersion [km / s]', fontsize=23)
plt.yscale('log')
plt.xlim([-1.0, 20.0])
```

```python
X_varySD = []
for i in surface_density_range:
    gmc.colDen = i
    #print(gmc.Tg)
    X_varySD.append(gmc.colDen/gmc.lineLum('co')[0]['intTB'])
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(surface_density_range, X_varySD, s=200)
plt.ylabel(r'X$_{CO}$ $[\frac{cm^{-2}}{K * km/s}]$', fontsize=25)
plt.xlabel('Surface Density [$cm^{-2}]$', fontsize=23)
plt.yscale('log')
plt.xscale('log')
#plt.xlim([-1.0, 20.0])
```
